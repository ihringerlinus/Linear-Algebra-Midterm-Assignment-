```python
#exercise 1e

import time
import random
import matplotlib.pyplot as plt

# Helper function to split a matrix into quarters
def split(matrix):
    n = len(matrix)
    mid = n // 2
    A11 = [row[:mid] for row in matrix[:mid]]
    A12 = [row[mid:] for row in matrix[:mid]]
    A21 = [row[:mid] for row in matrix[mid:]]
    A22 = [row[mid:] for row in matrix[mid:]]
    return A11, A12, A21, A22

# Helper function to add two matrices
def add_matrix(A, B):
    result = []
    for i in range(len(A)):
        row = []
        for j in range(len(A[0])):
            row.append(A[i][j] + B[i][j])
        result.append(row)
    return result

# Helper function to subtract two matrices
def subtract_matrix(A, B):
    result = []
    for i in range(len(A)):
        row = []
        for j in range(len(A[0])):
            row.append(A[i][j] - B[i][j])
        result.append(row)
    return result

# Recursive version of Strassen's algorithm
def strassen_recursive(A, B):
    n = len(A)
    if n == 1:
        return [[A[0][0] * B[0][0]]]
    else:
        A11, A12, A21, A22 = split(A)
        B11, B12, B21, B22 = split(B)

        M1 = strassen_recursive(add_matrix(A11, A22), add_matrix(B11, B22))
        M2 = strassen_recursive(add_matrix(A21, A22), B11)
        M3 = strassen_recursive(A11, subtract_matrix(B12, B22))
        M4 = strassen_recursive(A22, subtract_matrix(B21, B11))
        M5 = strassen_recursive(add_matrix(A11, A12), B22)
        M6 = strassen_recursive(subtract_matrix(A21, A11), add_matrix(B11, B12))
        M7 = strassen_recursive(subtract_matrix(A12, A22), add_matrix(B21, B22))

        C11 = add_matrix(subtract_matrix(add_matrix(M1, M4), M5), M7)
        C12 = add_matrix(M3, M5)
        C21 = add_matrix(M2, M4)
        C22 = add_matrix(subtract_matrix(add_matrix(M1, M3), M2), M6)

        # Combine the quarters into a full matrix
        C = []
        for _ in range(n):
            row = [0] * n
            C.append(row)
        for i in range(n // 2):
            for j in range(n // 2):
                C[i][j] = C11[i][j]
                C[i][j + n // 2] = C12[i][j]
                C[i + n // 2][j] = C21[i][j]
                C[i + n // 2][j + n // 2] = C22[i][j]
        return C

# Non-recursive version of Strassen's algorithm
def strassen_non_recursive(A, B):
    n = len(A)
    if n == 1:
        return [[A[0][0] * B[0][0]]]

    A11, A12, A21, A22 = split(A)
    B11, B12, B21, B22 = split(B)

    # Calculate M1 to M7 as per Strassen's formula without recursion
    M1 = matrix_multiply(add_matrix(A11, A22), add_matrix(B11, B22))
    M2 = matrix_multiply(add_matrix(A21, A22), B11)
    M3 = matrix_multiply(A11, subtract_matrix(B12, B22))
    M4 = matrix_multiply(A22, subtract_matrix(B21, B11))
    M5 = matrix_multiply(add_matrix(A11, A12), B22)
    M6 = matrix_multiply(subtract_matrix(A21, A11), add_matrix(B11, B12))
    M7 = matrix_multiply(subtract_matrix(A12, A22), add_matrix(B21, B22))

    # Calculate the four quadrants of the result matrix C
    C11 = add_matrix(subtract_matrix(add_matrix(M1, M4), M5), M7)
    C12 = add_matrix(M3, M5)
    C21 = add_matrix(M2, M4)
    C22 = add_matrix(subtract_matrix(add_matrix(M1, M3), M2), M6)

    # Combine the quadrants into the final result matrix C
    C = []
    for _ in range(n):
        row = [0] * n
        C.append(row)
    for i in range(n // 2):
        for j in range(n // 2):
            C[i][j] = C11[i][j]
            C[i][j + n // 2] = C12[i][j]
            C[i + n // 2][j] = C21[i][j]
            C[i + n // 2][j + n // 2] = C22[i][j]
    return C

# Naive matrix multiplication for submatrices
def matrix_multiply(A, B):
    n = len(A)
    result = []
    for _ in range(n):
        row = []
        for _ in range(n):
            row.append(0)
        result.append(row)
    for i in range(n):
        for j in range(n):
            result[i][j] = sum(A[i][k] * B[k][j] for k in range(n))
    return result

# Generate matrices with random numbers
def generate_matrix(size):
    matrix = []
    for _ in range(size):
        row = []
        for _ in range(size):
            row.append(random.randint(1, 10))
        matrix.append(row)
    return matrix

# Sizes of matrices to benchmark
matrix_sizes = [4, 8, 16, 32, 64, 128, 256, 512, 1024]  # Strassen works optimally on powers of 2
times_recursive = []
times_non_recursive = []

# Run the benchmark
for size in matrix_sizes:
    A = generate_matrix(size)
    B = generate_matrix(size)
    
    # Benchmark Recursive Strassen's algorithm
    start_time = time.time()
    strassen_recursive(A, B)
    end_time = time.time()
    duration_recursive = end_time - start_time
    times_recursive.append((size, duration_recursive))
    print(f"Matrix size {size}x{size} took {duration_recursive:.3f} seconds with Recursive Strassen's algorithm")

    # Benchmark Non-Recursive Strassen's algorithm
    start_time = time.time()
    strassen_non_recursive(A, B)
    end_time = time.time()
    duration_non_recursive = end_time - start_time
    times_non_recursive.append((size, duration_non_recursive))
    print(f"Matrix size {size}x{size} took {duration_non_recursive:.3f} seconds with Non-Recursive Strassen's algorithm")

# Plotting the results
sizes = [size for size, _ in times_recursive]
recursive_durations = [duration for _, duration in times_recursive]
non_recursive_durations = [duration for _, duration in times_non_recursive]

plt.plot(sizes, recursive_durations, label="Recursive Strassen's Algorithm", marker='o')
plt.plot(sizes, non_recursive_durations, label="Non-Recursive Strassen's Algorithm", marker='o')
plt.xlabel("Matrix Size (N x N)")
plt.ylabel("Time (seconds)")
plt.title("Performance Comparison of Recursive and Non-Recursive Strassen's Algorithm")
plt.legend()
plt.show()

# Print benchmark results
print("\nBenchmark Results (Recursive Strassen's Algorithm):")
for size, duration in times_recursive:
    print(f"Size {size}x{size}: {duration:.3f} seconds")

print("\nBenchmark Results (Non-Recursive Strassen's Algorithm):")
for size, duration in times_non_recursive:
    print(f"Size {size}x{size}: {duration:.3f} seconds")
```
